# Eigenface Method

```julia
k = 40
trial = 2

# load Data
path = "./HeadsAndTails/"
files = readdir(path)
n = 100
finalDim = (n,n)
fileInd = rand(1:size(files,1),n)
curfile = joinpath(path,files[1])
temp = convert(Array{Float64,2},convert(Image{Gray},load(curfile)))
temp = Images.imresize(temp,finalDim)
dim = size(temp)
D = zeros(prod(dim), n)
D[:,1] = vec(temp)
for i=2:n
    curfile = joinpath(path,files[fileInd[i]])
    temp = convert(Array{Float64,2},convert(Image{Gray},load(curfile)))
    temp = Images.imresize(temp,finalDim)
    D[:,i] = vec(temp)
end

x_bar = zeros(size(D,1),1)
for i =1:size(D,2)
    x_bar = x_bar + D[:,i]
end
x_bar = x_bar/size(D,2)
X_tilde = D - repmat(x_bar ,1,size(D,2))

covm = X_tilde'*X_tilde
U, S, V = svd(covm)
Vk = zeros(size(V,1),k)
Vk = V[:,1:k]
Uk = X_tilde*Vk
norms = diag(sqrt(abs(Uk'*Uk)))

A = Uk./ repmat(norms',size(Uk,1),1)

Px = zeros(k,size(D,2))
for i=1:n
    Px[:,i] = A'*X_tilde[:,i]
end
;
```

```julia
dimension = n*n
testim = convert(Array{Float64,2},Images.imresize(convert(Image{Gray},load("./Test/test1.jpg")), (500,500)))
m = round(Int,(500/n))
Y = zeros(dimension, round(Int,m^2))
counter = 1
for i=1:m
    for j=1:m
        rw = ((i-1)*n+1):(i*n)
        cl = ((j-1)*n+1):(j*n)
        Y[:,counter] = vec(testim[rw,cl])
        counter+=1
    end
end

Y_tilde = Y# - repmat(x_bar,1,size(Y,2))

Py = zeros(k,m^2)
for i=m^2
    Py[:,i] = A'*Y_tilde[:,i]
end

FF = zeros(m^2,n)
for i=1:m^2
    for j=1:n
        F = norm(Py[:,i]-Px[:,j])
        FF[i,j] = F;
    end
end
min_FF,index = findmin(FF)

y_ind = div(index,n)+1
grayim(reshape(Y[:,y_ind],(n,n))')
```
